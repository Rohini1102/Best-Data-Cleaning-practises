**Cuisine Distribution**

```python
b = pd.pivot_table(a,index='Cuisine_name',values='name_of_Dish', aggfunc='count')
a_sorted = b.sort_values('Cuisine_name', key=lambda x: x.map(b['name_of_Dish']), ascending=False)
print(a_sorted.head(10))
a_sorted.head(10).plot(kind='bar',figsize=(8,5),title='Top 10 Cuisines  types')
```

**Time Analysis**

```python
c = pd.pivot_table(a,index='Cuisine_name',values='Total_time',aggfunc='mean')
a_sorted = c.sort_values('Total_time',ascending=False)
print(a_sorted.head(10))
a_sorted.head(10).plot(kind='bar',figsize=(8,5),title='Top 10 Average Total time, Cuisines')
```

**Ingredient Analysis**

```python
ingredient_list = a['Ingredients_of_Dish'].str.split(',')
ingredients = [ingredient for sublist in ingredient_list for ingredient in sublist]
count=Counter(ingredients)
Top_ingredients = count.most_common(10)  #gives you a list of the 10 most common ingredients as tuples
print(Top_ingredients)
ingredient_names, ingredient_counts = zip(*Top_ingredients)  #"unzips" these tuples into two separate lists: ingredient_names (a list of ingredient names) and ingredient_counts (a list of corresponding counts).

plt.figure(figsize=(8,5))  # Adjust figure size as needed
plt.bar(ingredient_names, ingredient_counts)
plt.xlabel('Ingredients')
plt.ylabel('Count')
plt.title('Top 10 Most Common Ingredients')
plt.xticks(rotation=45,ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
